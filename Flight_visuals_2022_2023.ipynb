## 2023

```python
import pandas as pd
import requests
from io import StringIO
url = "https://group1-project4-flightdata2023.s3.us-east-2.amazonaws.com/2023.csv"

response = requests.get(url)

# Check if the request was successful
if response.status_code == 200:
    # Read the CSV content into a Pandas DataFrame
    pandas_df = pd.read_csv(StringIO(response.text))
    # Now 'pandas_df' is a Pandas DataFrame
    print(pandas_df.head())
else:
    print(f"Failed to fetch the CSV file. Status Code: {response.status_code}")
```

```python
import pandas as pd
import matplotlib.pyplot as plt


# Group by airport and count cancelled flights
cancelled_by_airport = pandas_df.groupby('ORIGIN')['CANCELLED'].sum().reset_index()

# Sort the airports by the number of cancelled flights in descending order
top_airports = cancelled_by_airport.sort_values(by='CANCELLED', ascending=False).head(20)

# Plot the top 20 airports
plt.figure(figsize=(10, 6))
plt.bar(top_airports['ORIGIN'], top_airports['CANCELLED'], color='#25316D')
plt.title('Top 20 Airports by Cancelled Flights 2023')
plt.xlabel('Airport Code')
plt.ylabel('Number of Cancelled Flights')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

```python
# Calculate the total number of flights and the number of cancelled flights
total_flights = len(pandas_df)
cancelled_flights = pandas_df['CANCELLED'].sum()

# Calculate the percentage of cancelled flights
cancelled_percentage = (cancelled_flights / total_flights) * 100

# Create a DataFrame for visualization
data = {'Category': ['Cancelled', 'Not Cancelled'], 'Count': [cancelled_flights, total_flights - cancelled_flights]}
df_visualization = pd.DataFrame(data)

# Plot a pie chart
plt.figure(figsize=(8, 8))
plt.pie(df_visualization['Count'], labels=df_visualization['Category'], autopct='%1.1f%%', startangle=90, colors=['red', '#25316D'])
plt.title('Flight Cancellation Rate 2023')
plt.show()

# Alternatively, plot a bar chart
plt.figure(figsize=(8, 6))
plt.bar(df_visualization['Category'], df_visualization['Count'], color=['red', '#25316D'])
plt.xlabel('Flight Status')
plt.ylabel('Number of Flights')
plt.title('Flight Cancellation Rate 2023')
plt.show()
```

```python
# Aggregate counts for each cancellation reason
cancellation_reasons = pandas_df[['DELAY_DUE_CARRIER', 'DELAY_DUE_WEATHER', 'DELAY_DUE_NAS', 'DELAY_DUE_SECURITY', 'DELAY_DUE_LATE_AIRCRAFT']]
reason_counts = cancellation_reasons.sum()

# Plot a bar chart
plt.figure(figsize=(10, 6))
reason_counts.plot(kind='bar', color='#25316D')
plt.xlabel('Cancellation Reasons')
plt.ylabel('Number of Cancellations')
plt.title('Reasons for Flight Cancellations 2023')
plt.xticks(rotation=45, ha='right')  # Rotate x-axis labels for better readability
plt.show()
```
